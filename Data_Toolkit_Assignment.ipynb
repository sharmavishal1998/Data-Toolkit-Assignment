```python
#Question1-Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each method.
#Answer-
import numpy as np

# Method 1: Using np.array()
array_1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Method 1:\n", array_1)

# Method 2: Using np.full() and reshaping or setting values manually
array_2 = np.full((2, 3), 0)  # Initialize with zeros
array_2[0, :] = [1, 2, 3]
array_2[1, :] = [4, 5, 6]
print("Method 2:\n", array_2)

# Method 3: Using np.zeros() and assigning values
array_3 = np.zeros((2, 3), dtype=int)
array_3[0] = [1, 2, 3]
array_3[1] = [4, 5, 6]
print("Method 3:\n", array_3)
```

```python
#Question3-Explain the following terms;
#the difference in np.array(),np.asarray() and np.asanyarray()
#the difference between deepcopy and shallowcopy
#Answer-
#np.array()-This function always creates a new array
#If the input is already an array (or array-like), it copies the data by default unless specified otherwise.
#You can control the behavior using the copy parameter (e.g., np.array(data, copy=False)).
#It also allows you to specify the data type of the array using the dtype argument
import numpy as np

# Example: Creates a new array
lst = [1, 2, 3]
arr = np.array(lst)


#np.asarray()-This function is used to convert input data to an array, but it does not create a copy if the input is already an array. It avoids copying to save memory and computation time.
#If the input is not an array, it will convert it to a NumPy array
# Example: No copy if input is already an array
arr = np.array([1, 2, 3])
arr_copy = np.asarray(arr)  # No new array created, just a reference to arr


#np.asanyarray()-Similar to np.asarray(), but it allows subclassing. This means that if you pass a subclass of an array (e.g., a matrix or masked array), it will return the same type rather than forcing it to be a regular ndarray.
# Example: Maintains subclass if possible
mat = np.matrix([[1, 2], [3, 4]])
arr = np.asanyarray(mat)  # arr will remain a matrix, not converted to ndarray

"""np.array(): Always creates a new array unless specified not to copy.
np.asarray(): Converts input to an array, avoids copying if input is already an array.
np.asanyarray(): Similar to np.asarray(), but preserves subclasses like matrix and masked_array."""

#Difference between deepcopy and shallowcopy
#Shallow copy-Copies only the outermost object
#Inner objects (like lists within lists) are shared between the original and the copy, so changes to mutable elements affect both.

#Deepcopy-Copies both the outer object and all nested objects recursively
#Inner objects are not shared, so changes to any part of the deep copy do not affect the original.

```

```python
#Question5-Create a Numpy array with random integers between 1 and 10 of shape(5,6),After creating the array perform the following operations:
#Extract all even integers from array.
#Extract all odd integers from array
#Answer-
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

# Extract all even integers from the array
even_integers = random_array[random_array % 2 == 0]

# Extract all odd integers from the array
odd_integers = random_array[random_array % 2 != 0]

print("Original Array:\n", random_array)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

```python
#Question06- Create a 3D numpy array of shape(3,3,3) containing random integers between 1 and 10 perform the following operations:
# Find the indices of the maximum values along each depth level (third axis).
#Perform element wise multiplication of between both array
#Answer-
import numpy as np

# Create a 3D array of shape (3, 3, 3) with random integers between 1 and 10
array_3D = np.random.randint(1, 11, (3, 3, 3))

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3D, axis=2)

# Perform element-wise multiplication of the array with itself
elementwise_multiplication = array_3D * array_3D

print("3D Array:\n", array_3D)
print("\nIndices of max values along each depth level (axis 2):\n", max_indices)
print("\nElement-wise multiplication:\n", elementwise_multiplication)
```

```python
#Question14-Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# Label the x-axis as 'Date'.
# Set the title of the plot as 'Temperature and Humidity Over Time'.
#Answer-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(low=20, high=35, size=100)  # Temperature between 20 and 35 degrees Celsius
humidity = np.random.uniform(low=30, high=80, size=100)      # Humidity between 30% and 80%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plotting
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
#Question17-Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
#Answer-
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
x = np.random.uniform(-10, 10, 100)  # x-coordinates
y = np.random.uniform(-10, 10, 100)  # y-coordinates

# Create a DataFrame for easier plotting
df = pd.DataFrame({'X': x, 'Y': y})

# Define a function to determine the quadrant
def get_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant I'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant II'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant III'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant IV'
    else:
        return 'On Axis'

# Apply the function to create a new column for quadrants
df['Quadrant'] = df.apply(get_quadrant, axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis', alpha=0.8)

# Add legend, labels, and title
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
#Question18-With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
#Answer-
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine wave values

# Create a new plot
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis',
           width=800, height=400)

# Add the sine wave line
p.line(x, y, line_width=2, color='blue', legend_label='y = sin(x)')

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
output_notebook()  # Use this line if you're running in a Jupyter notebook
show(p)
```

```python
#Question19-Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on theirvalues, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'
#Answer-
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare categorical data
categories = [f'Category {i}' for i in range(10)]  # 10 categories
values = np.random.randint(10, 100, size=len(categories))  # Random values for each category

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new bar chart
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           height=400, width=800)  # Changed to height and width

# Create a bar chart with color based on values
p.vbar(x='Category', top='Value', source=source,
        line_color='white', fill_color='blue')

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Customize the plot
p.xgrid.grid_line_color = None  # Remove grid lines for better visibility
p.y_range.start = 0  # Start y-axis at 0

# Show the plot
output_notebook()  # Use this line if you're running in a Jupyter notebook
show(p)

```

```python
#Question20-Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'
#Answer-
import plotly.graph_objects as go
import numpy as np

# Generate random dataset
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # Random y-values

# Create a line plot
fig = go.Figure()

# Add line plot to the figure
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes and set the title
fig.update_layout(title='Simple Line Plot',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis')

# Show the plot
fig.show()
```
